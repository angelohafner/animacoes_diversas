# Explicação do Código

O código que você forneceu é um exemplo de como criar uma animação simples usando o Matplotlib em Python. Aqui está uma explicação detalhada do que cada parte do código faz:

## Importações
- `import matplotlib.animation`: Importa o módulo de animação do Matplotlib, que permite criar animações.
- `import matplotlib.pyplot as plt`: Importa o módulo `pyplot` do Matplotlib, que é usado para criar figuras e gráficos.
- `import numpy as np`: Importa o NumPy, que é uma biblioteca para trabalhar com arrays e realizar cálculos matemáticos.

## Configurações
- `plt.rcParams["animation.html"] = "jshtml"`: Configura a saída da animação para ser exibida em notebooks Jupyter usando HTML e JavaScript.
- `plt.rcParams['figure.dpi'] = 150`: Define a resolução da figura em DPI (dots per inch), tornando a imagem mais nítida.
- `plt.ioff()`: Desliga o modo interativo do Matplotlib, o que impede que os gráficos sejam exibidos automaticamente enquanto o código é executado.

## Criação da Figura
- `fig = plt.figure()`: Cria uma nova figura onde a animação será desenhada.

## Função `animate(k)`
- Esta função é chamada em cada quadro da animação e recebe um parâmetro `k` que representa o número do quadro.
  - `plt.cla()`: Limpa o eixo atual para que o próximo quadro seja desenhado.
  - `t = 0.1*k`: Calcula o tempo `t` em função do quadro atual.
  - `plt.scatter(10*t, 50*t - 0.5*10*t**2)`: Desenha um ponto (`scatter plot`) em uma posição calculada por uma equação do movimento uniformemente acelerado (sem atrito) no eixo Y e movimento uniforme no eixo X.
  - `plt.xlim(0,120)`: Define os limites do eixo X entre 0 e 120.
  - `plt.ylim(0,150)`: Define os limites do eixo Y entre 0 e 150.

## Criação da Animação
- `matplotlib.animation.FuncAnimation(fig, animate, frames=100, interval=10)`: Cria a animação usando a função `animate`. A animação terá 100 quadros (`frames=100`), e cada quadro será exibido por 10 milissegundos (`interval=10`).

## Resumo
Este código cria uma animação de um ponto se movendo em um gráfico, onde a posição do ponto é determinada por uma equação de movimento. A animação mostra o ponto se movendo ao longo de uma trajetória parabólica (simulando um movimento projetado, como um objeto sendo lançado e caindo devido à gravidade) ao longo de 100 quadros.


```python
import matplotlib.animation as animation
import matplotlib.pyplot as plt
import numpy as np

# Configurações do Matplotlib
plt.rcParams["animation.html"] = "jshtml"  # Saída da animação em HTML e JavaScript
plt.rcParams['figure.dpi'] = 150  # Aumenta a resolução da figura

# Criação da figura para a animação
fig, ax = plt.subplots()
ax.set_xlim(0, 120)
ax.set_ylim(0, 150)

# Função de animação que será chamada a cada quadro
def funcao_animacao(k):
    ax.clear()  # Limpa o gráfico atual
    t = 0.1 * k  # Calcula o tempo baseado no quadro atual
    x = 10 * t  # Posição no eixo X (movimento uniforme)
    y = 50 * t - 0.5 * 10 * t**2  # Posição no eixo Y (movimento uniformemente acelerado)
    ax.scatter(x, y)  # Desenha o ponto na posição calculada
    ax.set_xlim(0, 120)  # Reaplica limites do eixo X
    ax.set_ylim(0, 150)  # Reaplica limites do eixo Y

# Criação da animação
ani = animation.FuncAnimation(fig=fig, func=funcao_animacao, frames=100, interval=100)

# Salva a animação como um arquivo de vídeo
ani.save('animacao.mp4', writer='ffmpeg', fps=100)

#Exibe a animação no ambiente de Jupyter Notebook (se necessário)
ani
```
